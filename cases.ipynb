```python
# case 10 a
features=(train[train.columns[0:12]])
features.corr()
```

```python
# case 10 b 
features.corr().unstack().sort_values(ascending = False)
def unique(list1):
    unique_list = []
    most_correlated=[]
    for x in list1:
        if x not in unique_list:
            unique_list.append(x)
        elif x not in most_correlated:
            most_correlated.append(x)
        else:
            continue
    return(most_correlated)
newList = unique(list(features.corr().unstack().sort_values(ascending = False)))
newList
```

```python
# case 11
from pandas.api.types import is_numeric_dtype

low = .05
high = .95
quant_train = train.quantile([low, high])
for name in list(train.columns):
    if is_numeric_dtype(train[name]):
        train = train[(train[name] > quant_train.loc[low, name]) & (train[name] < quant_train.loc[high, name])]
train
```
